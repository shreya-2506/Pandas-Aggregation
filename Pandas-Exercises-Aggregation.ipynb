# Aggregations

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
users = pd.read_csv('occupation.csv',sep='|') 
```

### Step 3. Assign it to a variable called users.

```python
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by = ['occupation'])[['age']].mean().add_prefix("MeanOf_")
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users['gender_new'] = np.where(users['gender']=='M',1,0)
male = users.groupby(by = ['occupation'])['gender_new'].sum()
male_ratio = (male/users['occupation'].value_counts())*100
male_ratio.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg([min,max]).rename(columns={"min":"MinOf_Age","max":"MaxOf_Age"})
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by = ['occupation','gender'])[['age']].mean().add_prefix("MeanOf_").reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
#get count per occupation, per gender
cnt_gender = users.groupby(['occupation', 'gender'])[['gender']].count()

#get count per occupation
cnt_total = users.groupby(['occupation'])[['gender']].count()

# Calculate percentage
(cnt_gender/cnt_total)*100
```
